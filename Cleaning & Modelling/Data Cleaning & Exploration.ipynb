```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(color_codes=True)

df = pd.read_csv('../data/kc_house_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Invetigating why sqft_basement is in object/string format
df['sqft_basement'].unique()
```

```python
# Seems like those that collated this dataframe was using a '?' as a placeholder for Null data.
df[df['sqft_basement'].str.contains("\?")]
```

```python
# Only 450 rows of data have this placeholder which is very small when compared to the total. 
# Rather than replace it with a mean of the column, because we don't know if these houses had basements at all
# the placeholder will be replaced with a 0
df['sqft_basement'] = df['sqft_basement'].replace("\?", 0, regex=True)
df['sqft_basement'] = df['sqft_basement'].astype(float)
```

```python
# Exploring how many houses are in the 'mansion' grade
(df['grade'] == 13).sum()
```

```python
# How many houses are the oldest in the dataframe
(df['yr_built'] == 1900).sum()
```

```python
# From describe(), it is shown that the sqft_lot has a very high standard deviation
sns.histplot(df['sqft_lot']);
```

```python
# Seeing if it's just an single outlier causing this skew
mean = df['sqft_lot'].mean()
std = df['sqft_lot'].std()
lot_3std = df[(df['sqft_lot'] >= mean + (3*std))]
lot_3std
```

```python
# Just under 350 rows above 3 standard deviations, so it's not errors
# These incredibly large lot sizes must be farming or agricultural land, such as Ranches
# This data is not appropriate for the business problem at hand, so they are removed
df = df[(df['sqft_lot'] <= mean + (3*std))]
```

```python
# Still skewed, but much more reasonable
sns.histplot(df['sqft_lot']);
```

```python
# yr_renovated is showing a min of 0, 
# which isn't correct data but it will show us how many houses were never renovated prior to being sold
(df['yr_renovated'] == 0).sum()
```

```python
df['yr_renovated'] = df['yr_renovated'].replace(0, np.NaN, regex=True)
df[~df['yr_renovated'].isnull()]
```

```python
# Only 731 houses were renovated
# Because the houses that are being constructed will be sold as new, 
# these houses are not applicable to the business problem, and are therefore removed
df = df[df['yr_renovated'].isnull()]
```

```python
# Before removing it from the dataframe, exploring the zipcode distributions
sns.histplot(df['zipcode']);
```

It's very skewed categorical data, with some zipcodes completely missing. Would need further data sources to investigate it's correlation to price.

```python
# Removing the non-architectural variables
df = df.drop(['id', 'date', 'condition', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'view', 'waterfront', 'yr_renovated', 'yr_built', 'yr_renovated', 'zipcode'], axis=1)
```

```python
df = df.reindex(sorted(df.columns), axis=1)
df.info()
```

```python
# bedrooms showed a max of 33, exploring if this is in line with it's other values
df['bedrooms'].unique()
```

```python
df[df['bedrooms'] == 33]
```

```python
# With only one value being this high, and the sqft_living not being much higher than other houses,
# it's apparent that this was an input error, and it most likely is meant to be a 3
df['bedrooms'] = df['bedrooms'].replace(33, 3)
```

```python
df['bedrooms'].unique()
```

```python
# For further understanding, checking the final cleaned dataframe for which variables correlate with price so far
price_corr = df.corr()[['price']].sort_values(by='price', ascending=False)

plt.figure(figsize=(4, 6))
heatmap = sns.heatmap(price_corr, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Features Correlating with Price', fontsize=14)
plt.savefig('../Images/Price_corr.png', bbox_inches='tight');
```
