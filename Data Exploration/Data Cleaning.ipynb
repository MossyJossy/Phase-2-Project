```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(color_codes=True)

df = pd.read_csv('data/kc_house_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['waterfront'] = df['waterfront'].astype('bool')
```

```python
df['sqft_basement'].unique()
```

```python
df[df['sqft_basement'].str.contains("\?")]
```

```python
df['sqft_basement'] = df['sqft_basement'].replace("\?", np.NaN, regex=True)
df['sqft_basement'] = df['sqft_basement'].astype(float)
```

```python
df[df['yr_renovated'] == 0]
```

```python
df['yr_renovated'] = df['yr_renovated'].replace(0, np.NaN, regex=True)
df[~df['yr_renovated'].isnull()]
```

```python
df = df[df['yr_renovated'].isnull()]
```

```python
df['date'].str[-4:]
```

```python
df['date'] = df['date'].str[-4:]
df['date'] = df['date'].astype(int)
sns.histplot(df['date']);
```

```python
df['house_age'] = (df['date'] - df['yr_built'])
```

```python
sns.histplot(df['house_age']);
```

```python
df[df['house_age'] <= 5]
```

```python
df = df.drop(['date', 'id', 'condition', 'yr_renovated', 'sqft_living15', 'sqft_lot15', 'view'], axis=1)
```

```python
df = df.reindex(sorted(df.columns), axis=1)
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(11, 7))
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))
heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Feature Correlation Heatmap', fontsize=16);
```

```python
price_corr = df.corr()[['price']].sort_values(by='price', ascending=False)

plt.figure(figsize=(4, 6))
heatmap = sns.heatmap(price_corr, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Features Correlating with Price', fontsize=14);
```

```python
df = df.drop('sqft_above', axis=1)
```

```python
df_new_house = df[df['house_age'] <= 5]
```

```python
df_new_house.info()
```

```python
price_corr = df_new_house.corr()[['price']].sort_values(by='price', ascending=False)

plt.figure(figsize=(4, 6))
heatmap = sns.heatmap(price_corr, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Features Correlating with Price', fontsize=14);
```
